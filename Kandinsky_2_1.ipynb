```python
import torch
from diffusers import StableDiffusionPipeline, DPMSolverMultistepScheduler

model_id = "stabilityai/stable-diffusion-2-1"

pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe.scheduler = DPMSolverMultistepScheduler.from_config(pipe.scheduler.config)
pipe = pipe.to("cuda")

#prompt = "a photo of an astronaut riding a horse on mars"
prompts = [
    "A surrealist photography of a futuristic city",
    "A Kandinsky-styled abstract painting with pink and purple elements",
    "An impressionist sunset over the sky",
    "Torn posters in Kandinsky style"
    "Jean Tinguely's abstract sculpture in the style of Kandinsky, geometric shapes, vibrant colors",
]

for i, prompt in enumerate(prompts):
    image = pipe(prompt, num_inference_steps=25).images[0]
    image.show()

#image = pipe(prompt).images[0]
image.save(f"promt{i+1}.png")
```
