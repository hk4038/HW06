```python
import torch
from diffusers import StableDiffusionPipeline, DPMSolverMultistepScheduler

model_id = "stabilityai/stable-diffusion-2-1"

# Use the DPMSolverMultistepScheduler (DPM-Solver++) scheduler here instead
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe.scheduler = DPMSolverMultistepScheduler.from_config(pipe.scheduler.config)
pipe = pipe.to("cuda")

prompts = [
    "Jean Tinguely's Homage to New York with An abstract colorful painting in the style of Kandinsky, geometric shapes, vibrant colors"
    "A cubist painting of a futuristic city",
    "A Kandinsky-inspired abstract artwork with blue and red elements",
    "An impressionist sunset over the ocean",
]

for i, prompt in enumerate(prompts):
    image = pipe(prompt, num_inference_steps=25).images[0]
    image.show()

    image.save(f"{i+1}.png")

#image.save("Tinguely_style.png")
```
